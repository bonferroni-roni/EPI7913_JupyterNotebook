## Add two new columns to present status and sex as categorical columns:

```r
# convert status to categorical
lung <- lung %>% 
    mutate(status.category = ifelse(status==1, "censored", "dead"),
           sex.category = ifelse(sex==1, "Male", "Female"))

# take a look!
head(lung)
```

Note that `status.category` and `sex.category` are character vectors now, you can also make them factors and specify the factor levels.

```r
summary(lung %>% select(status.category, sex.category))
```

```r
# make them factors 
lung$status.category <- factor(lung$status.category, levels = c("censored", "dead"))
lung$sex.category <- factor(lung$sex.category, levels = c("Male", "Female"))

# take a look
summary(lung %>% select(status.category, sex.category))
```

# Multivariate visualization: grouping

- I'll also use this as an example of how to arrange multiple plots.

```r
# Only run this cell if you haven't install the "ggpubr" package yet
install.packages("ggpubr")
```

```r
lung <- lung %>%
    mutate(time_cat = case_when(lung$time/365 <= 1 ~"<=1", 
                                lung$time/365 <= 2 ~"1-2",
                                TRUE ~ ">2"))
```

# Multivariate visualization: scatter plot of age vs. survival for the groups

```r
p1 <- ggplot(lung %>% filter(sex.category == "Male"), aes(time/365, age, color = time_cat, shape = time_cat)) +
        geom_point()+
        labs(title = "Male",
             x = "Survival Time (Years)", 
             y="Age")+
    theme(text = element_text(size = 14))

p2 <- ggplot(lung %>% filter(sex.category == "Female"), aes(time/365, age, color = time_cat, shape = time_cat)) +
        geom_point()+
        labs(title = "Female",
             x = "Survival Time (Years)", 
             y="Age")+
    theme(text = element_text(size = 14))

ggpubr::ggarrange(p1, p2, ncol = 2, common.legend = TRUE)
```

# Multivariate visualization: scatter plot of age vs. survival for the groups
## together

```r
# first group plot
ggplot(lung, aes(time/365, age, color = time_cat, shape = sex.category)) +
        geom_point()+
        labs(x = "Survival Time (Years)", 
             y="Age")+
    theme(text = element_text(size = 14))
```

# Multivariate visualization: grouping
## The easy way!

```r
# build the groups for every year of survival for either males or females
# by adding a new column into lung and creating new group labels
lung <- lung %>%
    mutate(new = case_when(sex.category=="Male" & lung$time/365 <= 1 ~"Group_m1", 
                           sex.category=="Male" & lung$time/365 <= 2 ~"Group_m2",
                           sex.category=="Male" & lung$time/365 <= 3 ~"Group_m3",
                           sex.category=="Female" & lung$time/365 <= 1 ~"Group_f1", 
                           sex.category=="Female" & lung$time/365 <= 2 ~"Group_f2",
                           sex.category=="Female" & lung$time/365 <= 3 ~"Group_f3"))
```

```r
head(lung)
```

# Multivariate visualization: scatter plot of age vs. survival for the groups
## together (easy)

```r
ggplot(lung, aes(time/365, age, color = new)) +
        geom_point()+
        labs(x = "Survival Time (Years)", 
             y="Age")+
    theme(text = element_text(size = 14))
```

# Multivariate visualization: estimation plot for multiple comparison

```r
multi.group <- 
  lung %>%
  dabest(new, time, 
         idx = list(c('Group_m1','Group_m2','Group_m3'), 
                    c('Group_f1','Group_f2','Group_f3')),
         paired = FALSE
        )

multi.group.mean_diff <- multi.group %>% mean_diff() 

plot(multi.group.mean_diff, color.column = sex)
```
