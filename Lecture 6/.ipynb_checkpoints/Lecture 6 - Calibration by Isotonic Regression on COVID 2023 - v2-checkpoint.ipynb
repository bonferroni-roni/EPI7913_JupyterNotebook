# Lecture 6 - Calibration by Isotonic Regression

## Load the COVID data set we used in the past

```r
df <- epi7913A::oncovid
df <- df[complete.cases(df),]
df$case_status <- as.factor(df$case_status)

set.seed(171)

head(df)
```

## Split the data into train, valid and test as 60/20/20 portions

```r
# split the data into train, valid and test as 60/20/20 portions
inds <- splitTools::partition(df$case_status, p = c(train = 0.6, valid = 0.2, test = 0.2))
```

## Balance the training data with both under- and over- sampling

```r
# balance the training data
df_train <- ROSE::ovun.sample(case_status ~., data=df[inds$train,], p=0.5, seed=11, method="both")$data

# check!
table(df_train$case_status)
```

## Train a random forest model with 100 trees to predict case_status and present classification performance on test data

```r
# build the andom forest rf
rf <- randomForest::randomForest(case_status ~., data= df_train, ntree=100)

# obtain predictions from the bag of tree models and average the results
p1 <- predict(rf, df[inds$test,])
caret::confusionMatrix(p1, df[inds$test,"case_status"])
```

## For easy of use, obtain predictions and outcome columns for both validation and test data

```r
# y.tr <- as.numeric(df[inds$train,1])-1
#pred.tr <- predict(rf, df[inds$train,], type='prob')

#y.vd <- as.numeric(df[inds$valid,1])-1
#pred.vd <- predict(rf, df[inds$valid,], type='prob')

#y.ts <- as.numeric(df[inds$test,1])-1
#pred.ts <- predict(rf, df[inds$test,], type='prob')


# extract labels for validation set
y.vd <- as.numeric(as.character(df[inds$valid,1]))

# obtain probability predictions for validation set
pred.vd <- predict(rf, df[inds$valid,], type='prob')

# extract labels for test set
y.ts <- as.numeric(as.character(df[inds$test,1]))

# obtain probability opredictions for test set
pred.ts <- predict(rf, df[inds$test,], type='prob')
```

## Remove duplicate prediction values of the validation set

```r
ids <- duplicated(pred.vd[,'1'])
pred.vd_unique <- pred.vd[!ids,]
y.vd_unique <- y.vd[!ids]
```

## Construct the isotonic regression model
### - train a logistic regression model on the validation dataset

```r
# train a logistic regression model on the cross validation dataset
m <- stats::isoreg(pred.vd_unique[,'1'], y.vd_unique)
```

### - plot the isotonic regression model of adjustment

```r
plot(m,plot.type = "row", main="Non-decreasing monotone isotonic regression model")
```

### To calculated the adjusted probabilities on the validation data set, we will use fit.isoreg() function.

### The *fit.isoreg* function below is defined for smoothing out the stepping function for better prediction -- technically, unnecessary!
### The source of this function is:
### https://www.analyticsvidhya.com/blog/2016/07/platt-scaling-isotonic-regression-minimize-logloss-error/

```r
# Source of this function is:
# https://www.analyticsvidhya.com/blog/2016/07/platt-scaling-isotonic-regression-minimize-logloss-error/
fit.isoreg <- function(iso, x0){
  o = iso$o
  if (is.null(o))
    o = 1:nrow(x)
  x = iso$x[o]
  y = iso$yf
  ind = cut(x0, x, labels = FALSE, include.lowest = TRUE)
  min.x <- min(x)
  max.x <- max(x)
  adjusted.knots <- iso$iKnots[c(1, which(iso$yf[iso$iKnots] > 0))]
  
  fits = sapply(seq(along = x0), function(i) {
    j = ind[i]

    # Handles the case where unseen data is outside range of the training data
    if (is.na(j)) {
      if (x0[i] > max.x) j <- length(x)
    else if (x0[i] < min.x) j <- 1
    }

    # Find the upper and lower parts of the step
    upper.step.n <- min(which(adjusted.knots > j))
    upper.step <- adjusted.knots[upper.step.n]
    lower.step <- ifelse(upper.step.n==1, 1, adjusted.knots[upper.step.n -1] )

    # Perform a liner interpolation between the start and end of the step
    denom <- x[upper.step] - x[lower.step]
    denom <- ifelse(denom == 0, 1, denom)
    val <- y[lower.step] + (y[upper.step] - y[lower.step]) * (x0[i] - x[lower.step]) / (denom)

    # Ensure we bound the probabilities to [0, 1]
    val <- ifelse(val > 1, max.x, val)
    val <- ifelse(val < 0, min.x, val)
    val <- ifelse(is.na(val), max.x, val) 
    val
  })
  fits
}
```

### - Obtain the calibrated probability scores for the validation data

```r
# perform the isotonic regression on the validation dataset
temp_df <- data.frame(pred.vd[,'1'],y.vd)
colnames(temp_df) <- c("predictions","labels")

results_adjusted_vd <- fit.isoreg(m, as.numeric(temp_df[,1]))
```

### - Obtain the calibrated probability scores for the test data

```r

tmp <- data.frame(pred.ts[,'1'],y.ts)
colnames(tmp) <- c("predictions","labels")

results_adjusted_ts <- fit.isoreg(m, as.numeric(tmp[,1]))
```

## Evaluate the calibration by Brier score
### There is a major improvement in Brier score when testing the probability predictions on the test data especially after the probability scores are adjusted.

```r
# calculate Brier score on testing with/without isotonic regression
bs_ts1 <- verification::brier(y.ts,pred.ts[,'1'])
bs_ts2 <- verification::brier(y.ts,results_adjusted_ts)

cat("Brier score on testing before abjustment:",bs_ts1$bs,"\n")
cat("Brier score on testing after abjustment:",bs_ts2$bs,"\n")
```

## Examine the Log Loss

### Adjusting the probability estimates (calibrating them) achieves the lowest log loss on test data and on verification data.

```r
# calculate Log Loss on testing without isotonic regression
v4 = ModelMetrics::logLoss(as.numeric(as.character(y.ts)),pred.ts[,'1'])
cat("Log Loss on testing before adjustment:", v4, "\n")

# calculate Log Loss on testing after isotonic regression
v5 = ModelMetrics::logLoss(as.numeric(as.character(y.ts)),results_adjusted_ts)
cat("Log Loss on testing after adjustment:", v5, "\n")
```

## Inspect the AUC of the model
### Adjusting the probability estimate to achieve better calibration does not affect teh area under the ROC curve (or the AUC) achieved by the prediction model.

```r
# calculate AUC on testing without isotonic regression
v4 = sdgm::auc(pred.ts[,'1'],as.numeric(as.character(y.ts)))
cat("AUC on testing before adjustment:", v4, "\n")

# calculate AUC on testing after isotonic regression
v5 = sdgm::auc(results_adjusted_ts,as.numeric(as.character(y.ts)))
cat("AUC on testing after adjustment:", v5, "\n")
```

## Plot the reliability plots (or calibration plots) **BEFORE** calibration adjustment

```r
# plotting reliability plots
A <- verification::verify(y.ts,pred.ts[,'1'], frcst.type = "prob", obs.type = "binary", 
                          thresholds=seq(0,1,by=0.2), bins=T)
verification::reliability.plot(A)
```

## Plot the reliability plots (or calibration plots) **AFTER** calibration adjustment

```r
B <- verification::verify(y.ts,results_adjusted_ts, frcst.type = "prob", obs.type = "binary",
                          thresholds=seq(0,1,by=0.2), bins=T)
verification::reliability.plot(B)
```

## For fun! Examine the Receiver Operating Characteristics (ROC) Curves before and after adjusting the probability scores -- no change at all.

```r
myROC <- pROC::roc(y.ts,pred.ts[,'1'], auc=T)
plot(1-myROC$specificities, myROC$sensitivities, type="o", xlab="False Positive Rate (1 - Specificity)", ylab="True Positive Rate (Sensitivity)",
     main="ROC Curve before calibration on Test Data")

myROC <- pROC::roc(y.ts,results_adjusted_ts, auc=T)
plot(1-myROC$specificities, myROC$sensitivities, type="o", xlab="False Positive Rate (1 - Specificity)", ylab="True Positive Rate (Sensitivity)", 
     main="ROC Curve after calibration on Test Data")
```
