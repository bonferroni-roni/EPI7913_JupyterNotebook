# Lecture 6 -  Patterns of Missingness on CCHS Data

## Load the CCHS data from epi7913A package

```r
library(magrittr)

full_data <- epi7913A::cchs_na 

head(full_data)
```

### Define the outcome variable

```r
# define the outcome variable
voutcome<-"CANHEARTbin"
```

### Determine the list of features and the outcome (dependant) variable

```r
# get names of all columns
tmp <- names(full_data)

# extract column names but exclude the outcome variable
feature_list <- tmp[tmp != voutcome]

# define the dependant variable (outcome)
dependent_variable <- voutcome
```

### Show the patterns of missingness between features and dependent variables

```r
# plot the missingness patterns and show the table of their frequencies
full_data %>% finalfit::missing_pattern(dependent_variable, feature_list)
```
